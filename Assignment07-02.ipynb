```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
from sklearn.preprocessing import scale 
```

```python
# Import Dataset
df_cri=pd.read_csv("/content/drive/MyDrive/Datasets/Clustering/crime_data.csv")
df_cri
```

```python
### Drop column
df_crime=df_cri.drop(["Unnamed: 0"],axis=1)
```

```python
df_crime.head()
```

```python
df_crime.shape
```

```python
df_crime.info()
```

#### THERE IS NO NULL VALUE

```python
df_crime.isnull().sum()
```

### Scaling the data

```python
df_crime_array=df_crime.values
```

```python
df_crime_norm=scale(df_crime_array)
```

```python
df_crime_norm
```

### Hierarchical Clustering

```python
#### create dendrogram{method="complete"}
dendrogram=sch.dendrogram(sch.linkage(df_crime_norm, method="complete"))
```

#### Create Clusters

```python
hc=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc=hc.fit_predict(df_crime_norm)
y_hc
```

```python
import sklearn
sklearn.metrics.silhouette_score(df_crime_norm,y_hc)
```

```python
hcrime=df_cri.copy()
hcrime["hcluster"]=y_hc
hcrime
```

```python
hcrime["hcluster"].value_counts()
```

```python
# create dendrogram{method="ward"}
dendrogram=sch.dendrogram(sch.linkage(df_crime_norm, method="ward"))
```

```python
#creating cluster
hc=AgglomerativeClustering(n_clusters=4,affinity="euclidean",linkage="ward")
```

```python
y_hc=hc.fit_predict(df_crime_norm)
y_hc
```

```python
import sklearn
sklearn.metrics.silhouette_score(df_crime_norm,y_hc)
```

```python
hcrime=df_cri.copy()
hcrime["hcluster"]=y_hc
hcrime
```

```python
hcrime["hcluster"].value_counts()
```

#### In second cluster n_cluster=4 and the silhouette_score=0.33 where linkage = ward.
#### so, the model is good compair to first one.

# KMeans Clustering

```python
from sklearn.cluster import KMeans
import  matplotlib.pyplot as plt
```

```python
sse=[]
k_rng =range(1,4)
for k in k_rng:
    km =KMeans(n_clusters=k)
    km.fit(df_crime_norm)
    sse.append(km.inertia_)
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

```python
model=KMeans(n_clusters=2,random_state=30).fit(df_crime_norm)
model.labels_ 
```

```python
#### Assign clusters to the data set
crimek=df_cri.copy()
crimek["kcluster"]=model.labels_ 
crimek.head()
```

```python
crimek["kcluster"].value_counts()
```

```python
sklearn.metrics.silhouette_score(df_crime_norm,model.labels_ )
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan=DBSCAN(eps=0.40,min_samples=5)
dbscan.fit(df_crime_norm)
```

```python
dbscan.labels_
```

#### All the dbscan.labels_ are -1 at eps=0.40 

```python
dbscan=DBSCAN(eps=0.70,min_samples=5)
dbscan.fit(df_crime_norm)
```
